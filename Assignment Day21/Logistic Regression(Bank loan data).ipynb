```python
y = data.loc[:,data.columns =="Personal Loan"]
x=data.loc[:, data.columns != 'Personal Loan']
```

```python
x.head()
```

```python
y.head()
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic=sm.Logit(y,x1)
```

```python
result=Logistic.fit()
```

```python
result.summary()
```

The p-values for most of the variables are smaller than 0.05, except five variables, therefore, we will remove them.

```python
df = data.drop(["ID","Age","Experience","ZIP Code","Mortgage"],axis=1)
df.head()
```

```python
x=df.loc[:, df.columns != 'Personal Loan']
y = df.iloc[:,4:5]
y.head()
# df.describe()
```

```python
x.head()
```

```python
Logistic=sm.Logit(y,x)
result=Logistic.fit()
print(result.summary())
```

Now All the p-Vlaues are less than 0.05 


We are building the Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

Result is conlusing that this model has 96% of accuracy !!!!

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

The result is telling us that we have 902+57 correct predictions and 33+8 incorrect predictions.
