```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Imorting Random forest algorithum</h2>

    For insampling of decision tree to find the important variable from dataset


```python
from sklearn.ensemble import RandomForestClassifier

rfModel=RandomForestClassifier(warm_start=True,oob_score=True)

```

```python
data1.dtypes
```

# Changing the categorical features to numerical

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

data1["Attrition"]=le.fit_transform(data1["Attrition"])
data1["BusinessTravel"]=le.fit_transform(data1["BusinessTravel"])
data1["Department"]=le.fit_transform(data1["Department"])
data1["Gender"]=le.fit_transform(data1["Gender"])
data1["JobRole"]=le.fit_transform(data1["JobRole"])
data1["MaritalStatus"]=le.fit_transform(data1["MaritalStatus"])
data1["Over18"]=le.fit_transform(data1["Over18"])
data1["EducationField"]=le.fit_transform(data1["EducationField"])

data1.head()
```

defining The features

```python
data1.columns
```

```python
features = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField','Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'PercentSalaryHike','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
rfModel.fit(X=data1[features],y=data1["Attrition"]);
```

```python
print("Obb Accuracy for this model is :  ")
print(rfModel.oob_score_);
```

```python
for feature,imp in zip(features,rfModel.feature_importances_):
    print(feature,"     :",imp)
```

# From the RF-Model we can see that factors effecting to Attrition is :  

Age,MonthlyIncome: more than 10%

TotalWorkingYears,YearsAtCompany,PercentSalaryHike  :   more than 5%

we can draw Decision tree for this Features  :

```python
# For decision tree importing required libraries  :


from sklearn import tree


treeModel = tree.DecisionTreeClassifier()

impPredictors=pd.DataFrame([data1["Age"],data1["MonthlyIncome"]]).T;

treeModel.fit(X=impPredictors,y=data1["Attrition"]);
```

```python
with open('Dtree.dot','w') as f:  #decision tree always store in .dot format
    f=tree.export_graphviz(treeModel,feature_names=["Age","PercentSalaryHike"],out_file=f);
```

```python
treeModel.score(X=impPredictors,y=data1["Attrition"])
```
